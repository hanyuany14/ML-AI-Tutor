## Matplotlib Basics

### Figure

Figure 是圖表的整體窗口或畫布。它包含所有的圖形、文本、標題和軸。您可以將 Figure 理解為整個圖表的容器。



```python
import matplotlib.pyplot as plt

# 創建一個新圖表（Figure）
fig = plt.figure()

# 添加子圖（Ax subplot to the figure）
ax = fig.add_subplot(111)

# 繪製數據
ax.plot([1, 2, 3], [4, 5, 6])

# 顯示圖表
plt.show()
```

### Axes

Axes 是圖表中的一個區域，它包含了數據、刻度、標籤、標題等。每個 Axes 對應於一個子圖，它是 Figure 的一部分。



```python
import matplotlib.pyplot as plt

# 創建一個新圖表（Figure）
fig, ax = plt.subplots()

# 繪製數據
ax.plot([1, 2, 3], [4, 5, 6])

# 設置標題和標籤
ax.set_title('Example Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 創建一個包含兩個子圖的圖表
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# 繪製第一個子圖的數據
ax1.plot([1, 2, 3], [4, 5, 6], marker='o', linestyle='--', color='b')
ax1.set_title("Example Plot 1")
ax1.set_xlabel("X Axis 1")
ax1.set_ylabel("Y Axis 1")

# 繪製第二個子圖的數據
ax2.plot([1, 2, 3], [6, 5, 4], marker='s', linestyle='-', color='r')
ax2.set_title("Example Plot 2")
ax2.set_xlabel("X Axis 2")
ax2.set_ylabel("Y Axis 2")

# 調整子圖之間的間距
plt.tight_layout()

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 創建一個包含兩個子圖的圖表
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

# 繪製第一個子圖的數據
ax1.plot([1, 2, 3], [4, 5, 6], marker='o', linestyle='--', color='b')
ax1.set_title("Example Plot 1")
ax1.set_xlabel("X Axis 1")
ax1.set_ylabel("Y Axis 1")

# 繪製第二個子圖的數據
ax2.plot([1, 2, 3], [6, 5, 4], marker='s', linestyle='-', color='r')
ax2.set_title("Example Plot 2")
ax2.set_xlabel("X Axis 2")
ax2.set_ylabel("Y Axis 2")

# 調整子圖之間的間距
plt.tight_layout()

# 顯示圖表
plt.show()
```

### Axis

Axis 是圖表中的一個單獨的軸。Axes 包含了兩個（或更多）Axis，即 x 軸和 y 軸。Axis 控制著數據的範圍、刻度和標籤。



```python
import matplotlib.pyplot as plt

# 創建一個包含兩個子圖的圖表
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# 繪製第一個子圖的數據
ax1.plot([1, 2, 3], [4, 5, 6], marker='o', linestyle='--', color='b')
ax1.set_title("Subplot 1")
ax1.set_xlabel("X Axis 1")
ax1.set_ylabel("Y Axis 1")
ax1.set_xlim(0, 4)
ax1.set_ylim(3, 7)

# 繪製第二個子圖的數據
ax2.plot([1, 2, 3], [6, 5, 4], marker='s', linestyle='-', color='r')
ax2.set_title("Subplot 2")
ax2.set_xlabel("X Axis 2")
ax2.set_ylabel("Y Axis 2")
ax2.set_xlim(0, 4)
ax2.set_ylim(3, 7)

# 調整子圖之間的間距
plt.tight_layout()

# 顯示圖表
plt.show()
```

### Subplot

Subplot 是在一個 Figure 中創建多個 Axes 的一種方式。它允許您在同一個 Figure 中佈局多個子圖。



Sub_plots: 如何快速生成大量 plots ?
如果我們這一次要建立好幾個子圖（axes），那我們不就要新增一整排的變數axe_x？因此衍生的更便利的用法 plt.subplots() 可以使用（和前面的方式差了一個 ”s”），因為 axes 可以直接像二維陣列一樣做索引，也可以一起設定不同子圖間的關連。



```python
import matplotlib.pyplot as plt

# 創建一個包含 2x2 子圖的圖表（Figure）
fig, axs = plt.subplots(2, 2)

# 在每個子圖中繪製數據
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].plot([1, 2, 3], [6, 5, 4])
axs[1, 0].plot([1, 2, 3], [5, 4, 6])
axs[1, 1].plot([1, 2, 3], [4, 6, 5])

# 顯示圖表
plt.show()
```

```python
import matplotlib.pyplot as plt

# 創建一個包含 4x1 子圖的圖表（Figure）
fig, axs = plt.subplots(4, 1, figsize=(8, 12))

# 在每個子圖中繪製數據
axs[0].plot([1, 2, 3], [4, 5, 6])
axs[0].set_title("Plot 1")

axs[1].plot([1, 2, 3], [6, 5, 4])
axs[1].set_title("Plot 2")

axs[2].plot([1, 2, 3], [5, 4, 6])
axs[2].set_title("Plot 3")

axs[3].plot([1, 2, 3], [4, 6, 5])
axs[3].set_title("Plot 4")

# 調整子圖之間的間距
plt.tight_layout()

# 顯示圖表
plt.show()
```

## 基本繪圖

### 1. 導入 Matplotlib

首先，我們需要導入必要的庫。我們通常導入 Matplotlib 的 pyplot 模塊並將其別名為 plt 以便使用。

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. 創建一個簡單的折線圖

**適用數據和應用場域**

- 時間序列數據：折線圖非常適合顯示隨時間變化的數據，例如股價、氣溫變化等。
- 趨勢分析：用於識別趨勢或模式，如銷售趨勢、流量趨勢等。
- 連續數據：適合連續測量的數據，例如生產過程中的測量數據。

**參數說明**

- x: X 軸數據
- y: Y 軸數據
- label: 圖例中的標籤
- color: 線條顏色
- linestyle: 線條樣式（'-' 實線, '--' 虛線, '-.' 點劃線, ':' 點線）
- linewidth: 線條寬度

```python
# 創建一些數據
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 創建折線圖
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)

# 添加標題和標籤
plt.title('折線圖示例')
plt.xlabel('X 軸')
plt.ylabel('Y 軸')

# 添加圖例
plt.legend()

# 顯示圖表
plt.show()
```

## 3. 自定義圖表

我們可以以各種方式自定義圖表，例如更改線條樣式、顏色、添加網格線等等。



```python
# 創建一些數據
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 創建一個具有自定義線條的圖表
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-.')

# 添加標題和標籤
plt.title('自定義折線圖')
plt.xlabel('X 軸')
plt.ylabel('Y 軸')

# 添加圖例
plt.legend()

# 添加網格線
plt.grid(True)

# 顯示圖表
plt.show()
```

## 4. 創建不同類型的圖表

Matplotlib 支持各種類型的圖表。以下是幾個示例：



**柱狀圖**

**適用數據和應用場域**

- 類別數據：適合顯示不同類別之間的比較，例如銷售量、人口數量等。
- 頻率分佈：用於顯示各類別的頻率或計數。
- 調查數據：適合展示問卷調查結果，例如滿意度調查等。

**參數說明**

- x: X 軸數據
- height: Y 軸數據
- color: 柱狀顏色
- width: 柱狀寬度
- align: 柱狀對齊方式（'center' 或 'edge'）


```python
# 數據
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# 創建柱狀圖
plt.bar(categories, values, color='green', width=0.5, align='center')

# 添加標題和標籤
plt.title('柱狀圖示例')
plt.xlabel('類別')
plt.ylabel('數值')

# 顯示圖表
plt.show()
```

**散點圖**

**適用數據和應用場域**

- 雙變量數據：用於展示兩個變量之間的關係，例如身高和體重、年齡和收入等。
- 相關性分析：適合分析兩個變量之間的相關性。
- 異常值檢測：可以幫助識別數據中的異常值或離群點。

**參數說明**

- x: X 軸數據
- y: Y 軸數據
- color: 點的顏色
- marker: 點的標記樣式（如 '.' 圓點, 'o' 圓圈, '^' 三角形等）
- s: 點的大小

```python
# 創建一些數據
x = np.random.rand(50)
y = np.random.rand(50)

# 創建散點圖
plt.scatter(x, y, color='red', marker='o', s=100)

# 添加標題和標籤
plt.title('散點圖示例')
plt.xlabel('X 軸')
plt.ylabel('Y 軸')

# 顯示圖表
plt.show()
```

**直方圖**


**適用數據和應用場域**

- 連續數據分佈：用於顯示連續數據的分佈情況，例如測量數據的分佈。
- 頻率分析：適合分析數據的頻率分佈，例如成績分佈、收入分佈等。
- 數據範圍：幫助理解數據的範圍和集中趨勢。

**參數說明**

- x: 數據
- bins: 柱狀數量
- edgecolor: 柱狀邊緣顏色
- alpha: 透明度（0 到 1 之間）

```python
# 創建一些數據
data = np.random.randn(1000)

# 創建直方圖
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)

# 添加標題和標籤
plt.title('直方圖示例')
plt.xlabel('數值')
plt.ylabel('頻率')

# 顯示圖表
plt.show()
```

**箱型圖**

**適用數據和應用場域**

- 數據分佈：用於顯示數據的分佈情況，特別是四分位數和中位數。
- 異常值檢測：適合識別數據中的異常值。
- 多組數據比較：可用於比較多組數據的分佈，例如不同班級的成績分佈。


**參數說明**

- x: 數據
- patch_artist: 是否填充盒子（True 或 False）
- boxprops: 盒子的屬性（如顏色）
- whiskerprops: 鬚鬚的屬性（如顏色）

```python
# 創建一些數據
data = [np.random.randn(100) for _ in range(4)]

# 創建箱型圖
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'), whiskerprops=dict(color='blue'))

# 添加標題和標籤
plt.title('箱型圖示例')
plt.xlabel('類別')
plt.ylabel('數值')

# 顯示圖表
plt.show()
```

**圓餅圖**

**適用數據和應用場域**

- 類別數據：適合顯示各類別的比例，例如市場份額、人口比例等。
- 構成分析：用於分析整體中各部分的構成。
- 百分比表示：適合展示數據的百分比分佈。


**參數說明**

- x: 數據
- labels: 標籤
- colors: 顏色
- autopct: 自動標註百分比格式
- startangle: 起始角度
- shadow: 是否有陰影（True 或 False）

```python
# 數據
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# 創建圓餅圖
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)

# 添加標題
plt.title('圓餅圖示例')

# 顯示圖表
plt.axis('equal')  # 保持圓餅圖為圓形
plt.show()
```
