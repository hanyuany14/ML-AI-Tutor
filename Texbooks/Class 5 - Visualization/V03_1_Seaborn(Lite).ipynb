## Seaborn Basics

#### Basic plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 加載範例數據集
tips = sns.load_dataset("tips")
```

使用 Seaborn 創建基本的圖表通常包括以下步驟：

設置圖表風格：使用 sns.set_style() 設置圖表風格，如 sns.set_style('whitegrid')。

繪製圖表：使用不同的 Seaborn 函數來繪製圖表，如 sns.barplot()、sns.histplot()、sns.scatterplot() 等。

顯示圖表：使用 plt.show() 或直接在 Jupyter Notebook 中顯示。

```python
sns.set_style('whitegrid')
sns.histplot(data=tips, x='total_bill', bins=10, kde=True) # 這裡就是設定不同的表樣式
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

#### Make an interative plot

```python
import plotly.express as px
import seaborn as sns

# 載入示例數據集
tips = sns.load_dataset('tips')

# 使用 Plotly Express 創建互動散點圖
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', hover_data=['size', 'day'])

# 設置標題和軸標籤
fig.update_layout(
    title='Tips vs Total Bill',
    xaxis_title='Total Bill ($)',
    yaxis_title='Tip ($)'
)

# 顯示圖表
fig.show()
```

### Set the Theme

```python

# 設置主題和調色板
sns.set_theme(style="darkgrid")
sns.set_palette("pastel")

# 畫一個簡單的圖表
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
plt.title("Scatterplot with Darkgrid Theme")
plt.show()

# 切換到 whitegrid 主題
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
plt.title("Scatterplot with Whitegrid Theme")
plt.show()

# 切換到 ticks 主題
sns.set_theme(style="ticks")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
plt.title("Scatterplot with Ticks Theme")
plt.show()
```

### 分佈圖 (Distribution Plots)


#### 直方圖（Histogram）


- **介紹**：直方圖是一種用於顯示數據分佈的圖表，通過將數據分成若干區間（bins），並計算每個區間的數據頻率來繪製。
- **特性**：可以清楚地顯示數據的集中趨勢、偏態和異常值。
- **應用場域**：適用於數值型數據的分佈分析，例如人口統計數據、考試成績等。

```python
# 直方圖
sns.histplot(tips['total_bill'], kde=True)
plt.title("Total Bill Histogram with KDE")
plt.show()
```

#### 核密度圖（KDE Plot）


- **介紹**：核密度圖是估計數據的概率密度函數的圖表，使用核密度估計（KDE）來平滑數據分佈。
- **特性**：相比直方圖，KDE圖更平滑，可以更好地顯示數據的連續性。
- **應用場域**：適用於數據分佈的平滑估計，例如金融數據、醫學數據等。

```python
# 核密度圖
sns.kdeplot(tips['total_bill'], shade=True)
plt.title("Total Bill KDE Plot")
plt.show()
```

#### 成對關係圖（Pair Plot）



- **介紹**：成對關係圖是一種展示多變量數據集中每對變量之間關係的圖表。
- **特性**：可以顯示多變量數據集中每對變量的散佈圖和直方圖，並且可以顯示變量之間的相關性。
- **應用場域**：適用於多變量數據分析，例如機器學習中的特徵工程、探索性數據分析等。

```python
# 成對關係圖
sns.pairplot(iris, hue='species')
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

### 分類圖 (Categorical Plots)


#### 條形圖（Bar Plot）

- **介紹**：條形圖用於顯示分類數據的頻率或計數。
- **特性**：每個條形的高度或長度表示該分類的數據量。
- **應用場域**：適用於分類數據的頻率分析，例如銷售數據、調查數據等。


```python
# 條形圖
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Average Total Bill by Day")
plt.show()
```

#### 箱線圖（Box Plot）


- **介紹**：箱線圖是一種顯示數據分佈的摘要圖表，通過五個數據點（最小值、第一四分位數、中位數、第三四分位數、最大值）來描述數據的分佈情況。
- **特性**：可以顯示數據的集中趨勢、變異性和異常值。
- **應用場域**：適用於數據分佈的對比分析，例如測量數據、成績數據等。

```python
# 箱線圖
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Total Bill Distribution by Day")
plt.show()
```

#### 點圖（Strip Plot）



- **介紹**：點圖是一種顯示分類數據的圖表，每個數據點在分類軸上以點的形式表示。
- **特性**：可以顯示每個分類的數據分佈和密度。
- **應用場域**：適用於顯示數據的個體分佈，例如實驗數據、測量數據等。

```python
# 點圖
sns.stripplot(x='day', y='total_bill', data=tips)
plt.title("Total Bill Strip Plot by Day")
plt.show()
```

### 回歸圖 (Regression Plots)


#### 線性回歸圖（Linear Regression Plot）


- **介紹**：線性回歸圖顯示數據點和最佳擬合直線，用於描述變量之間的線性關係。
- **特性**：可以顯示數據的趨勢和關聯程度。
- **應用場域**：適用於探索和建模變量之間的線性關係，例如經濟數據、社會科學數據等。

```python
# 線性回歸圖
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs Tip with Linear Regression")
plt.show()
```

#### 殘差圖（Residual Plot）


- **介紹**：殘差圖顯示回歸模型的殘差，用於檢查模型的擬合效果。
- **特性**：可以顯示殘差的分佈和模型的偏差。
- **應用場域**：適用於檢查回歸模型的適用性和診斷模型問題，例如經濟數據、社會科學數據等。

```python
# 殘差圖
sns.residplot(x='total_bill', y='tip', data=tips)
plt.title("Residual Plot of Total Bill and Tip")
plt.show()
```

###  矩陣圖 (Matrix Plots)


#### 熱圖（Heatmap）


- **介紹**：熱圖是一種顯示數據矩陣中數值大小的圖表，用顏色表示數值大小。
- **特性**：可以顯示數據矩陣的模式和趨勢。
- **應用場域**：適用於顯示相關矩陣、混淆矩陣等，例如機器學習、數據分析等。

在計算相關矩陣時，需要將非數值類型的數據排除或轉換。為了解決這個問題，可以使用select_dtypes方法來選擇數值類型的列進行相關矩陣的計算。



```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加載範例數據集
tips = sns.load_dataset("tips")

# 選擇數值類型的列
numeric_tips = tips.select_dtypes(include='number')

# 計算相關矩陣
corr_matrix = numeric_tips.corr()

# 繪製熱圖
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Heatmap of Tips Dataset Correlation Matrix")
plt.show()
```

### 聚類圖（Clustermap）

#### 介紹：
聚類圖（Clustermap）是一種結合了熱圖和層次聚類（hierarchical clustering）的圖表，用於展示數據矩陣的模式和聚類結果。通過將相似的數據分組，可以揭示數據中的內在結構和模式。

#### 特性：
- **顯示數據矩陣的模式**：使用顏色來表示數據矩陣中的數值大小。
- **層次聚類**：將相似的數據分組，生成一棵樹狀圖（dendrogram），顯示數據之間的關聯和距離。
- **雙重聚類**：可以對行和列同時進行聚類分析。

#### 應用場域：
- **基因表達數據**：在生物學和醫學中，用於顯示基因表達的聚類結果。
- **社會網絡分析**：用於顯示社會網絡中節點的關聯和聚類。
- **市場研究**：用於顯示客戶行為數據的模式和聚類。

### 如何分析聚類圖：
1. **觀察顏色模式**：聚類圖中顏色的變化可以顯示數據矩陣中的模式，例如高值和低值的聚集區域。
2. **分析樹狀圖（dendrogram）**：樹狀圖顯示了數據的層次結構，可以觀察數據分組的層次和相似性。
3. **比較行和列的聚類**：通過雙重聚類，可以同時比較行和列之間的關聯性。





```python
import seaborn as sns
import matplotlib.pyplot as plt

# 加載範例數據集
flights = sns.load_dataset("flights")

# 透過透視表格將數據轉換為適合的格式
flights_pivot = flights.pivot(index="month", columns="year",values="passengers")

# 繪製聚類圖
sns.clustermap(flights_pivot, cmap="coolwarm", standard_scale=1)
plt.title("Clustermap of Flights Dataset")
plt.show()
```



**分析聚類圖的步驟**

1. **觀察顏色模式**：
   - 深藍色表示較低的值（較少的乘客數量）。
   - 深紅色表示較高的值（較多的乘客數量）。
   - 可以看到，圖表的下半部分（夏季月份）主要是紅色，這表明這些月份乘客數量較多。上半部分（冬季月份）主要是藍色，乘客數量較少。

2. **分析樹狀圖（dendrogram）**：
   - 樹狀圖顯示了行（月份）和列（年份）之間的聚類結構。
   - 可以看到哪些月份的乘客數量模式相似，它們在樹狀圖中會更靠近。
   - 同樣，年份之間的聚類結構也能顯示哪些年份的乘客數量模式相似。

3. **雙重聚類比較**：
   - 橫向的樹狀圖（行聚類）顯示了月份之間的相似性。比如，7月、8月和9月被聚在一起，表示這些月份的乘客數量模式相似。
   - 縱向的樹狀圖（列聚類）顯示了年份之間的相似性。比如，1957年和1958年被聚在一起，表示這些年份的乘客數量模式相似。

**具體分析示例**

1. **季節性趨勢**：
   - 圖中的顏色變化顯示了明顯的季節性趨勢。夏季月份（如6月、7月、8月、9月）乘客數量顯著高於冬季月份（如1月、2月、3月、4月）。

2. **年份間的比較**：
   - 1957年和1958年的乘客數量模式相似，可以看到它們在樹狀圖中被聚在一起。
   - 類似地，1950年和1951年也被聚在一起，顯示了這些年份的乘客數量模式相似。

3. **異常數據點**：
   - 可以檢查是否有任何異常的數據點或模式。例如，如果某個月份在大多數年份中顯示出相似的顏色模式，但在某一年中顯示出不同的顏色，這可能表示該年的異常情況。
