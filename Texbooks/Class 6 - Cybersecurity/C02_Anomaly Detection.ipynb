## 重要概念

Anomaly detection refers to the problem of finding patterns in data that do not conform to expected behavior. These nonconforming patterns are often referred to as anomalies, outliers, discordant observations, exceptions, aberrations, surprises, peculiarities, or contaminants in different application domains.





a wide variety of applications such as fraud detection for credit cards, insurance, or health care, intrusion detection for cyber-security, fault detection in safety critical systems, and military surveillance for enemy activities.



Anomalies are patterns in data that do not conform to a well defined notion of normal behavior.



Challenges is to define a region representing normal behavior and declare any observation in the data that does not belong to this normal region as an anomaly
- Defining a normal region
- defining normal behavior
- normal behavior keeps evolving
-  The exact notion of an anomaly is different for different application domains.(異常的判斷值不一)
- Availability of labeled data
- noise

### Type of Anomaly

異常可以分為以下三種類別：

1. **點異常 (Point Anomalies)**：
   - **定義**：相對於其餘數據而言，單個數據實例被認為是異常的。
   - **例子**：在信用卡欺詐檢測中，某筆交易的花費金額遠高於該人正常花費範圍。

2. **上下文異常 (Contextual Anomalies aka conditional anomaly)**：
   - **定義**：在特定上下文中被認為是異常的數據實例，但在其他情況下則不是。
   - **上下文屬性**：用於確定實例上下文（或鄰域）的屬性，例如空間數據中的經度和緯度，或時間序列數據中的時間。
   - **行為屬性**：定義實例非上下文特徵的屬性，例如空間數據集中某地點的降雨量。
   - **例子**：在時間序列數據中，某個時間點的溫度（行為屬性）在冬季（上下文屬性）是正常的，但在夏季可能是異常的。








3. **集體異常 (Collective Anomalies)**：
   - **定義**：一組相關數據實例相對於整個數據集是異常的，即使單個實例本身可能不是異常。
   - **例子**：在心電圖（ECG）輸出中，一段持續時間異常低的數值，表明可能存在健康問題。

### Data labels



1. **監督異常檢測（Supervised Anomaly Detection）**：
   - **定義**：假設有一個包含正常和異常標籤的訓練數據集。
   - **方法**：構建一個預測模型來區分正常和異常類別，並使用該模型來分類新的數據實例。
   - **挑戰**：
     - 異常實例通常比正常實例少，導致類別分佈不平衡的問題。
     - 獲得準確且具代表性的異常標籤非常困難。
   - **技術**：一些技術通過向正常數據集中注入人工異常來獲取標記的訓練數據集。

2. **半監督異常檢測（Semisupervised Anomaly Detection）**：
   - **定義**：假設訓練數據集中只有正常類別的標籤。
   - **方法**：構建一個模型來表示正常行為，並使用該模型來識別測試數據中的異常。
   - **適用性**：比監督技術更廣泛，因為它不需要異常類別的標籤。
   - **挑戰**：僅在異常場景（如事故）不易建模時適用。

3. **無監督異常檢測（Unsupervised Anomaly Detection）**：
   - **定義**：不需要訓練數據，因此應用最廣泛。
   - **方法**：假設正常實例在測試數據中遠多於異常實例，根據這一假設來檢測異常。
   - **挑戰**：如果這一假設不成立，則這類技術容易產生較高的誤報率。

- **標籤需求**：
  - **監督異常檢測**：需要標記的正常和異常實例。
  - **半監督異常檢測**：只需要標記的正常實例。
  - **無監督異常檢測**：不需要任何標記的訓練數據。

- **適用性**：
  - **監督異常檢測**：適用於能夠獲得充足標記數據的情況，但受限於異常數據的稀缺性和高成本。
  - **半監督異常檢測**：適用於異常場景不易建模的情況，更加廣泛適用。
  - **無監督異常檢測**：最為廣泛適用，尤其是在標記數據缺乏的情況下，但假設正常實例占多數。

- **挑戰**：
  - **監督異常檢測**：處理類別不平衡和獲取準確標記數據的挑戰。
  - **半監督異常檢測**：建模正常行為的準確性。
  - **無監督異常檢測**：依賴於正常實例占多數的假設，否則誤報率高。

### CLASSIFICATION BASED ANOMALY DETECTION TECHNIQUES

image.png



根據訓練階段可用的標籤，基於分類的異常檢測技術可以分為兩大類：多類分類異常檢測技術（Multi-class Classification Based Anomaly Detection Techniques）和單類分類異常檢測技術（One-class Classification Based Anomaly Detection Techniques）。

1. **多類分類異常檢測技術（Multi-class Classification Based Anomaly Detection Techniques）**：
   - **定義**：假設訓練數據包含屬於多個正常類別的標記實例。
   - **方法**：這些技術訓練一個分類器來區分每個正常類別和其他類別。如果測試實例沒有被任何分類器分類為正常，則該實例被認為是異常的。
   - **特點**：有些技術會對分類器的預測結果附加一個置信度分數。如果沒有分類器對測試實例的分類有足夠的置信度，則該實例被宣告為異常。

2. **單類分類異常檢測技術（One-class Classification Based Anomaly Detection Techniques）**：
   - **定義**：假設所有訓練實例只有一個類別標記。
   - **方法**：這些技術使用單類分類算法來學習正常實例周圍的區分邊界。例如，單類支持向量機（One-class SVMs）、單類核費舍判別分析（One-class Kernel Fisher Discriminants）。任何不在學習邊界內的測試實例都被宣告為異常。




- **標籤需求**：
  - **多類分類異常檢測技術**：需要多個正常類別的標記實例。
  - **單類分類異常檢測技術**：只需要一個正常類別的標記實例。

- **分類方法**：
  - **多類分類異常檢測技術**：訓練分類器來區分每個正常類別和其他類別。
  - **單類分類異常檢測技術**：使用單類分類算法來學習正常實例周圍的區分邊界。

- **異常判定**：
  - **多類分類異常檢測技術**：如果測試實例沒有被任何分類器分類為正常，則該實例被認為是異常的。
  - **單類分類異常檢測技術**：任何不在學習邊界內的測試實例都被宣告為異常。

- **置信度**：
  - **多類分類異常檢測技術**：有些技術會附加置信度分數，如果沒有分類器有足夠的置信度，則實例被宣告為異常。
  - **單類分類異常檢測技術**：通常不涉及置信度分數，只根據是否落在邊界內來判定。

簡言之，多類分類異常檢測技術適用於需要區分多個正常類別的情況，而單類分類異常檢測技術適用於只有一個正常類別的情況。兩者在標籤需求、分類方法和異常判定標準上存在差異。

### 不同模型如何判斷異常值？(挑出較為常見與主要的方式)



-  CLASSIFICATION BASED ANOMALY DETECTION TECHNIQUES: Assumption. A classifier that can distinguish between normal and anomalous classes
can be learned in the given feature space.If none of the classifiers are confident in classifying the test instance as normal, the instance is declared to be anomalous.


Neural Network:
First, a neural network is trained on the normal training data to learn the different normal classes. Second, each test instance is provided as an input to the neural network. If the network accepts the test input, it is normal

- NEAREST NEIGHBOR-BASED: Normal data instances occur in dense neighborhoods, while anomalies occur far from their closest neighbors.(1) techniques that use the distance of a data instance to its kth nearest neighbor as the anomaly score;
(2) techniques that compute the relative density of each data instance to compute its anomaly score.
 A key advantage of nearest neighbor-based techniques is that they are unsupervised in nature and do not make any assumptions regarding the generative distribution for the data. Instead, they are purely data driven.

- CLUSTERING-BASED ANOMALY DETECTION TECHNIQUES: Clustering-based anomaly detection techniques can be grouped into three categories.

Assumption. Normal data instances belong to a cluster in the data, while anomalies do not belong to any cluster.

Assumption. Normal data instances lie close to their closest cluster centroid, while anomalies are far away from their closest cluster centroid.

Assumption. Normal data instances belong to large and dense clusters, while anoma- lies either belong to small or sparse clusters.

- STATISTICAL ANOMALY DETECTION: Assumption. Normal data instances occur in high probability regions of a stochastic model, while anomalies occur in the low probability regions of the stochastic model.

這篇文章《Anomaly Detection: A Survey》是由明尼蘇達大學的 Varun Chandola、Arindam Banerjee 和 Vipin Kumar 撰寫的。文章提供了異常檢測（Anomaly Detection）領域的一個結構化和全面的綜述，涵蓋了不同研究領域和應用領域內的技術。

1. **異常檢測的定義和重要性**：
   - 異常檢測是識別不符合預期行為的數據模式的問題，這些模式通常被稱為異常、異常值、異常觀察、例外、驚喜、特異或污染物。
   - 異常檢測在欺詐檢測、入侵檢測、故障檢測和軍事監控等應用中有廣泛的應用。

2. **異常檢測技術的分類**：
   - 文章將現有技術根據採用的方法分為不同類別，並對每個類別進行詳細討論，包括關鍵假設、基本技術、變體、優缺點和計算複雜度。
   - 主要技術包括基於分類、聚類、最近鄰、統計、信息理論和光譜的方法。

3. **挑戰**：
   - 定義正常行為的區域並不容易，尤其是在邊界不明確的情況下。
   - 惡意行為者通常會調整其行為以使其看起來正常。
   - 正常行為可能會隨時間演變，當前的正常行為模型可能在未來並不適用。

4. **異常的類型**：
   - **點異常**：單個數據點異常。
   - **上下文異常**：在特定上下文中異常的數據點。
   - **集體異常**：數據點集合異常，但單個數據點可能不異常。

5. **異常檢測的輸出**：
   - **得分**：為每個測試數據點分配異常得分。
   - **標籤**：為每個測試數據點分配正常或異常標籤。

6. **應用領域**：
   - **入侵檢測**：包括基於主機和網絡的入侵檢測系統。
   - **欺詐檢測**：信用卡欺詐、手機欺詐和保險詐欺。
   - **醫療和公共衛生**：病患紀錄和疾病爆發檢測。
   - **工業損害檢測**：機械單元和結構損害檢測。
   - **圖像處理**：包括衛星圖像、手寫數字識別和視頻監控。
   - **文本數據中的異常檢測**：新話題或事件檢測。
   - **傳感器網絡**：檢測傳感器故障或入侵。

7. **技術詳細討論**：
   - **基於分類的技術**：包括神經網絡、貝葉斯網絡、支持向量機和基於規則的方法。

這篇文章旨在提供對異常檢測領域研究的不同方向的更好理解，並展示在一個領域開發的技術如何應用於其他領域。它還識別了每種技術的優點和缺點，並討論了它們在實際應用中的計算複雜性。

無標籤（無監督學習）的異常檢測技術旨在不依賴預先標記的數據來識別異常。在這種情況下，我們假設數據是從某個未知分布 \( F(x) \) 中取樣出來的，並通過最大似然估計（Maximum Likelihood Estimation, MLE）來找到使得數據可能性最大的參數。

### 解釋

1. **數據分布假設**：
   - 我們假設數據 \( x \) 是從一個分布 \( F(x) \) 中取樣出來的。

2. **最大似然估計**：
   - 我們需要找到一組參數 \( \theta \)，使得在這組參數下，分布 \( F(x) \) 產生出這組數據的可能性最大。這個過程稱為最大似然估計。

3. **高斯分布模型**：
   - 假設我們使用高斯分布（正態分布）來建模數據。高斯分布由均值 \( \mu \) 和協方差矩陣 \( \Sigma \) 參數化。

4. **最大似然計算公式 \( L \)**：
   - 最大似然函數 \( L \) 表示在給定參數 \( \mu \) 和 \( \Sigma \) 下，數據 \( x \) 的可能性。公式如下：
     \[
     L(\mu, \Sigma) = \prod_{i=1}^{n} \frac{1}{(2\pi)^{d/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (x_i - \mu)^T \Sigma^{-1} (x_i - \mu)\right)
     \]
   - 其中，\( n \) 是數據點的數量，\( d \) 是數據的維度，\( x_i \) 是第 \( i \) 個數據點。

5. **參數估計**：
   - 通過最大化似然函數 \( L \)，找到最優參數 \( \mu \) 和 \( \Sigma \)。這可以通過數學優化技術來實現，如梯度下降或期望最大化算法（EM算法）。

6. **異常檢測**：
   - 一旦確定了最優參數 \( \mu \) 和 \( \Sigma \)，我們可以計算每個數據點的似然值。對於新的數據點 \( x \)，如果其似然值低於某個閾值，則該數據點被認為是異常。
   - 具體做法是計算數據點與高斯分布均值的距離，如果距離超過一定標準（例如3個標準差），則視為異常。

### 總結

無監督異常檢測通過估計數據的潛在分布參數，來識別那些與大多數數據點顯著不同的數據點。使用高斯分布進行建模並找到最大似然估計參數，可以幫助我們有效地檢測數據中的異常點，這種方法在無需標記數據的情況下特別有用。


簡單來說，就是找到一組最佳參數，套用在數學公式中，會讓數據中的正常數據相似度最高。當一組新的數據進來後，可以套用我們設定好的數學公式，得到一個分數，我們會預期不正常的數據分數會越低（代表越不相似），正常的數據分數應該要高（代表越相似）
