## 模型是什麼？What is model?


* A **toy model** is a deliberately simplistic model with many details removed while preserving important and necessary characteristics so that it can be used to explain something concisely.
  * The 'best' model of a subject is exactly the subject itself.
  * However, most of the time, a subject is too complex that we are not able to (or we do not know how to) describe it.

* A **machine learning model** can be a mathematical representation of a real-world relationship of data.

## AI 領域的關鍵字

* Table is a straightforward way to present data, and it usually includes
  * $n$ data points
  * each data point is represented by $m$ features, $x_1$ to $x_m$
  * the entire features can be represented as a vector $X$
  * it may have a specical column -- a target $y$
* A model $M$ is trying to represent the entire data (including $X$ and $y$).
  * As mentioned, the best way to represent the data is to keep the entire table.
  * However, the number of data points and/or the number of features might be large. It is not efficient to do so.
  * Now, we are trying to find a more concise way to represent the data. And more, if there is indeed a concise (mathematical) way to represent the $n$ data points, that means, the data points have some hidden relations (between instances and/or between $X$ and $y$), so that a concise representation is possbile.

* We consider a data table, such that
  * has a **traget** $y$
  * all **features** $X$
* A **model** $M: X \rightarrow y$ can be found by using the existing $n$ data points, a.k.a. **training data**.
  * Generally, $X$ and $y$ could be numeric or categorical values.
    * Different types of data and target have different algorithms to create a model.
    * If target $y$ is numeric, the model is called **regression model**.
    * If target $y$ is categorical, the model is called **classification model**.
  * Theoratically, there should be *lots of models* can be used to correctly represent the $n$ data points. However, most of the time, we want a concise one and can accept some small errors (it is a trade-off).
    * That means, maybe not all $n$ data points are represented.
    * Maybe $y$ value has an error.
  * As mentioned, there midght be a ground-truth function that can generate the training data (from a hidden distribution), $y = f(X)$. These $n$ data points can be viewed as some samples (observations) from the distribution.
  * A model can be viewed as an approximate function to represent the data in this way: $\hat{y} = g(X)$.
    * $\hat{y}$ is the **prediction** value of the model $g()$ while observing $X$.
    * $g()$ is an approximation of $f()$.

* If a model,
  * has $X$ and $y$, we can learn the relation between $X$ and $y$ by different algorithms. These algorithms belong to **supervised learning algorithm**.
  * has only $X$, we can learn the relation between these $n$ data points in $X$. Such algorithms belong to **unsupervised learning algorithm**.

## 人工智慧（AI）與機器學習（ML）的定義與區別


### 1. 人工智慧（AI, Artificial Intelligence）
- **定義**：人工智慧是一門研究和設計智能機器的技術，使其能模仿人類的思考、學習和解決問題的能力。它包括了從簡單的自動化工具到擁有高度智能和決策能力的系統。
- **範圍**：包括了從狹義AI（解決特定問題，如圖中的窄AI）到廣義AI（具有與人類相似的認知能力）的所有範疇。

### 2. 機器學習（ML, Machine Learning）
- **定義**：機器學習是人工智慧的一個分支，著重於開發能夠從數據中學習的算法，不需要明確的指令即可做出決策或預測。
- **範圍**：機器學習是達到AI的手段之一，專注於利用算法自動提取和學習數據的模式和特徵。

### 3. 神經網絡（NN, Neural Networks）
- **定義**：神經網絡是由大量的節點（或稱神經元）組成的網絡，模擬人腦的神經元網絡來處理信息。每個神經元與其他神經元通過連接權重互動。
- **範圍**：神經網絡是實現深度學習的結構基礎，可以用於各種機器學習任務，包括分類、回歸等。

### 4. 深度學習（DL, Deep Learning）
- **定義**：深度學習是機器學習的一種特殊形式，它使用深層的（多隱藏層的）神經網絡來學習數據的高層次抽象特徵。
- **範圍**：深度學習特別適合於處理大量複雜數據，如圖像和語音識別等領域，因為它能自動識別和學習數據中的複雜模式。

這些技術層層遞進，每一層都是基於前一層的技術發展而來，共同推動著人工智慧領域的進步。

```python
import numpy as np

# 定義 Sigmoid 激活函數
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 定義一個簡單的兩層神經網絡
class SimpleNeuralNetwork:
    def __init__(self):
        # 初始化權重和偏差
        self.input_size = 3
        self.hidden_size = 4
        self.output_size = 2

        # 隨機初始化權重和偏差
        self.W1 = np.random.randn(self.input_size, self.hidden_size)
        self.b1 = np.random.randn(self.hidden_size)
        self.W2 = np.random.randn(self.hidden_size, self.output_size)
        self.b2 = np.random.randn(self.output_size)

    def forward(self, X):
        # 第一層傳遞
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = sigmoid(self.z1)

        # 第二層傳遞
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = sigmoid(self.z2)

        return self.a2

# 創建一個簡單的神經網絡實例
nn = SimpleNeuralNetwork()

# 定義一個輸入向量
X = np.array([0.5, 0.1, 0.4])

# 前向傳遞
output = nn.forward(X)

print("Input Vector:", X)
print("Hidden Layer Output:", nn.a1)
print("Output Vector:", output)
```

## AI的主要分支

### 機器學習：使用統計技術使機器通過數據進行學習。
### 深度學習：一種特殊的機器學習方法，使用多層次的神經網絡模擬人類大腦處理資訊。
### 自然語言處理（NLP）：使機器能夠理解和生成人類語言的技術。
### 計算機視覺（Computer Vision）
**定義**：使機器能夠解讀和理解視覺信息（如圖像和視頻）的技術。
**應用**：圖像分類、物體檢測、面部識別、自動駕駛等。

### 強化學習（Reinforcement Learning）
**定義**：通過與環境的互動，機器學習如何在各種情況下做出決策以最大化某個獎勵信號。
**應用**：遊戲AI、機器人控制、自動駕駛等。

### 知識表示與推理（Knowledge Representation and Reasoning, KRR）
**定義**：研究如何以結構化的方式表示知識並推理出新的信息。
**應用**：專家系統、知識圖譜、語義網絡等。

### 規劃與調度（Planning and Scheduling）
**定義**：研究如何為機器制定行動計劃以達到目標。
**應用**：機器人導航、自動駕駛、供應鏈管理等。

### 機器感知（Machine Perception）
**定義**：使機器能夠從各種感知數據（如視覺、聽覺、觸覺）中獲取信息。
**應用**：語音識別、視覺感知、環境感知等。

### 人機交互（Human-Computer Interaction, HCI）
**定義**：研究如何設計和實現人與機器之間的交互界面。
**應用**：語音助手、智能家居、虛擬現實（VR）和增強現實（AR）等。

### 自然語言生成（Natural Language Generation, NLG）
**定義**：使機器能夠生成自然語言文本的技術。
**應用**：自動寫作、聊天機器人、數據報告生成等。


## 強AI（Strong AI） vs 弱AI（Weak AI）




美國哲學家約翰．瑟爾 (John Searle) 便提出了「強人工智慧」(Strong A.I.)和「弱人工智慧」(Weak A.I.) 的分類。



弱人工智慧（Weak Artificial Intelligence）：
透過電腦與軟體模擬「部分」的人類智慧與行為，像是只會下圍棋、只會自動導航、只會醫療影像判斷、只會掃地......等特定功能的AI，它們通常被應用在解決特定領域問題，因此也有人稱之為 Narrow AI、Applied AI，也是目前最常見的人工智慧。



強人工智慧（Strong Artificial Intelligence）：
期望電腦能執行所有人類的工作，除了具有認知能力之外，還能夠推理、自學、溝通甚至擁有自我意識，並綜合這些能力來面對未知的情境，可以因應不同狀況調整自身能力來實踐特定目標。

因為擁有非常廣泛的能力，面對各種情境都通用，因此也有人稱之為 Full AI、AGI（Artificial General Intelligence），這目前大概只有在電影裡看的到。

超人工智慧（Super Artificial Intelligence）：
有專家把最終階段定義為「AI在任何領域都超越人類大腦，包括科學、創新、通識與社交等技能」，這個階段已經跨過所謂的「科技奇異（Technological Singularity）」，AI可以自己創造出比自己更聰明、更強大的東西，創造出一個我們人類完全無法理解的世界。（這大概還非常遙遠吧）

目前支撐 AI 世界的依然是弱 AI:

簡單來說，從大量的數據資料裡，結合統計模型來對未來事件做出預測，正式目前機器學習領域裡，重要的發展方向，舉凡像是Gmail的自動過濾垃圾郵件、Who's Call的自動過濾騷擾電話等等，這些都已經反應到"弱人工智慧"在我們現在的社會裡，的確是可實現的。

・Siri（語音識別）：

Siri 是一個廣泛應用於蘋果產品的語音助手，它使用了深度學習算法進行語音識別，包括卷積神經網絡（CNN）（註1）和 循環神經網絡（RNN）。這些技術讓 Siri 能夠理解各種口音、語言，並能與用戶自然地進行交流。




・Google Photos（圖像識別）：

Google Photos 是一個圖片存儲和管理平台，擅長對圖片進行分類、識別和檢索。它運用了卷積神經網絡（CNN）進行圖像識別，能自動將照片按照人物、地點、事件等標籤進行整理，方便用戶找到所需的照片。




・ChatGPT（自然語言處理）：

ChatGPT 是一個自然語言處理模型，擅長對話、翻譯、文本生成等。它運用了 Transformer 架構和大量的語料訓練，使其具有更強大的理解和生成能力。ChatGPT 能與人類用戶進行自然的對話交流，並提供有趣、有幫助的建議。尤其近期的 GPT-4 發表後，不僅擁有更強的語言和理解能力，更可對圖像產生回應！

## 機器學習的定義

機器學習是一門使機器能夠從數據中學習規律並做出決策的科學，不需要明確程式化地指令。



## 深度學習簡介

- **什麼是深度學習？**：深度學習是一種通過多層神經網絡進行特徵抽取和轉換的機器學習技術，每層只根據前一層的資訊進行處理。
- **神經網絡的基本結構**：一般包括輸入層、多個隱藏層和輸出層。每層由多個神經元組成，神經元之間通過權重連接，通過激活函數處理信號。

舉例來說，像雅量裡面提到的，假設我們今天拿到了一張圖去做類神經分析，那他會去拆解裡面所有的圖像特徵，並用神經元來確認每一個形狀、字母、符號、尺寸等等，透過權重跟高度訓練的推測，可以讓系統有87分得把握來推斷他是一張稿紙、剩下13分的機率來推測他是一塊綠豆糕，最後透過網路把正確的結果回傳給神經網路做修正。



當然過程中會出現大量的錯誤答案，但這也是一個絕佳的訓練機會，透過成千上萬，數以百萬計的圖像去做校正，來校正每一個神經元的輸入權重，不斷的滾動來取得精準的正確答案，讓圖片僅管颳風下雨、打雷起霧、或是馬賽克等等，都可以明確的辨認圖片裡的綠色物體，到底是稿紙還是綠豆糕。

僅管現在把這樣的理論可以輕描淡寫的敘述，但在資源較匱乏的過去十年前，不論是硬體或是資源上，都無法去做這樣大量的計算資料，其中一個主因是因為大家都習慣把計算丟給電腦的CPU去做處理，但這樣透過加權分析的複雜計算，會用到大量的矩陣，而顯卡中的GPU才是最適合大量處理矩陣運算的硬體，但過去GPU不像CPU擁有編譯器 (Compiler) 這樣的設計。



Google 的 AlphaGo 學習圍棋遊戲遊戲，並且反覆跟自己比賽來調整神經網路的比重，訓練它進行圍棋比賽，讓每一步棋下得更為精準，這些都是實際上深度學習的應用。
