# Optimizer

## Introduction

優化器其實就是在做梯度下降時會使用的優化算法，可以理解成滑雪選手對於指南針的提示有不同的戰術

在機器學習和深度學習的領域中，優化器（Optimizer） 是用於調整模型參數（如神經網絡的權重和偏置）以最小化損失函數（也稱為目標函數或成本函數）的算法。優化器的目的是提高模型在訓練數據和測試數據上的性能。優化器使用梯度信息來指導模型參數的更新，從而加速模型的收斂速度。

## Optimizer 有哪些？

1. 隨機梯度下降（Stochastic Gradient Descent, SGD）

**概念**：
- 就像爬山一樣，想要找到山頂（最小化損失），但每次只看附近一小步的坡度。
- 每次使用一小部分數據（批量）來更新模型參數。

**白話說明**：
- 想像你在黑暗中爬山，每次只能用手電筒照亮你周圍的一小塊地方，你根據這塊地方的坡度來決定往哪走一步。



2. Momentum

**概念**：
- Momentum（動量）是一種優化技術，用於加速梯度下降算法的收斂速度。它通過引入過去梯度的加權平均來平滑梯度更新。Momentum 是「運動量」的意思，此優化器為模擬物理動量的概念，在同方向的維度上學習速度會變快，方向改變的時候學習速度會變慢。


- 它會考慮前幾次更新的方向，並在該方向上加速。


**白話說明**：
- 想像你在爬山，每次走一步的時候，你會考慮到之前走的方向，這樣如果你一直朝著正確的方向走，你會越走越快。

3. AdaGrad

**概念**：
- AdaGrad（Adaptive Gradient Algorithm）是一種自適應學習率的優化方法。它會為每個參數分配不同的學習率，根據參數歷史梯度的大小來調整學習率。
- 對於Optimizer來說，learning rate(學習率) η 相當的重要，太小會花費太多時間學習，太大有可能會造成overfitting，無法正確學習，前面幾種Optimizer的學習率 η，都為固定值，而AdaGrad就是會依照梯度去調整 learning rate η 的優化器，Ada對我來說就是Adaptive的意思



**白話說明**：
- 想像你在爬山，每次走一步的時候，你會根據之前每一步的陡峭程度來調整你的步伐。如果之前的路很陡，你會小心翼翼地走（小步），如果之前的路很平緩，你會大步向前。



4. Adam優化器

**概念**：
- Adam 結合了動量（Momentum）和自適應學習率（AdaGrad）的優點。
- 它會記住之前步驟的更新方向（動量），並調整每個參數的學習率（自適應學習率），使得學習率會隨著時間自動變小。

**白話說明**：
- 想像你在爬山，你不僅記住了你之前走過的路（動量），還會根據你在不同地方的進展速度來調整每一步的大小（自適應學習率），讓你更快地找到山頂。



5. RMSprop

**概念**：
- RMSprop（Root Mean Square Propagation）是對 AdaGrad 的改進。它通過引入衰減因子來防止學習率變得過小。
- 它使用指數加權移動平均來計算梯度平方的平均值。

**白話說明**：
- 想像你在爬山，每次走一步的時候，你會根據最近一段時間的陡峭程度來調整你的步伐，而不僅僅是根據所有歷史步伐。






6. LookAhead Optimizer

**概念**：
- LookAhead 會在每N步後，根據這N步的累積方向更新一次模型參數，從而平滑和穩定訓練過程。

**白話說明**：
- 想像你在爬山，你每走N步就停下來，回顧一下你走過的路，然後根據這段路的總體方向做出下一步的決策。

7. LARS（Layer-wise Adaptive Rate Scaling）

**概念**：
- LARS 針對每一層網絡的參數調整學習率，特別適用於訓練非常大的模型。

**白話說明**：
- 想像你在一個非常大的山脈裡爬山，不同的山峰有不同的坡度，你根據每個山峰的特點來決定每一步的大小。

8. 梯度中心化（Gradient Centralization）

**概念**：
- 將梯度向量的均值歸零，這有助於提高模型的訓練穩定性和速度。

**白話說明**：
- 想像你在黑暗中爬山，每次都會先調整你的手電筒，使得它的光線集中在你腳下的中心點，這樣你每一步都能更精確地走向山頂。

9. 自動混和精度訓練（Automatic Mixed Precision Training）

**概念**：
- 在訓練中同時使用低精度（如半精度浮點數）和高精度（如單精度浮點數）的計算，這可以加速訓練並減少內存使用。

**白話說明**：
- 想像你在爬山，有時候用大步（低精度計算）快速前進，有時候用小步（高精度計算）來做精細的調整，這樣你既能快又能準確地到達山頂。


- **Momentum**：在每次更新中考慮前幾次更新的方向，讓你走得更快更穩定。
- **AdaGrad**：根據歷史梯度調整學習率，讓步伐根據地形自適應。
- **RMSprop**：對 AdaGrad 的改進，引入衰減因子來防止學習率過小。


### 差異


學習率調整

- **固定學習率**：如基本的SGD，學習率在訓練過程中不變。

- **自適應學習率**：如Adagrad、RMSProp、Adam等，學習率會根據過去的梯度信息自適應地調整。

- **動態學習率調度**：一些優化器支持基於迭代次數或損失變化調整學習率。

收斂速度

- **收斂速度較慢**：如SGD，特別是在高曲率或震盪的損失表面。

- **收斂速度較快**：如Adam、Nadam，特別適合於深層神經網絡。

適應性和穩定性

- **適應性強**：如Adam和RMSProp，能夠處理非平穩的目標函數，並自適應調整學習率。

- **穩定性高**：有些優化器在面對噪聲和非凸問題時具有更高的穩定性，如動量和Nadam。

### 如何選擇

做實驗就對了



1. **模型類型和結構**
   - 對於較淺層或簡單模型，SGD或Momentum通常就足夠。
   - 對於深層神經網絡，Adam或其變種（如Nadam、AdamW）通常效果較好。

2. **數據集性質**
   - 稀疏數據：Adagrad可能更適合，因為它能自適應地調整學習率。
   - 大量數據：RMSProp和Adam可以提供更穩定和快速的收斂。

3. **訓練時間和計算資源**
   - 如果計算資源有限，SGD和Momentum由於計算簡單，可能是更好的選擇。
   - 如果希望加快收斂或處理大規模數據，Adam和Nadam是較好的選擇。

4. **對超參數的敏感度**
   - 有些優化器（如Adam）對學習率等超參數的設定不太敏感，適合快速試驗。

5. **過擬合問題**
   - 使用像AdamW這樣的優化器，可以提供更好的正則化效果，減少過擬合。

6. **問題的性質**
   - 非凸問題：使用帶有動量的優化器（如Momentum, Adam, Nadam）能有效避免陷入局部最小值。
   - 鞍點問題：使用動量或自適應學習率的優化器能更好地避開鞍點。
